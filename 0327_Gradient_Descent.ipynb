경사하강법 만들기

```python
import numpy as np
import matplotlib.pyplot as plt

# Learning Rate(학습률)
lr_list = [0.001, 0.1, 0.3, 0.7]

def get_derivative(lr_list):
  # 초기 가중치
  weight_old = 2
  # 가중치의 목록
  derivative = [weight_old]

  # 손실함수 : y = x^2
  y = [weight_old ** 2]

  # epochs = 10 : 밤복학습 횟수
  for i in range(1, 10):
    # 손실함수에 대한 기울기(손실함수 미분)
    dev_value = 2 * weight_old

    # 가중치 업데이트
    weight_new = weight_old - lr * dev_value
    weight_old = weight_new

    # 업데이트한 값을 저장
    # 가중치 업데이트 저장
    derivative.append(weight_old)
    #손실값 업데이트 저장
    y.append(weight_old ** 2)

  return derivative, y

# 데이터 만들기
x = np.linspace(-2, 2, 50)
x_square = [i**2 for i in x]

# 그래프 그리기
fig = plt.figure(figsize=(12,7))

for i, lr in enumerate(lr_list):
  derivative, y = get_derivative(lr)
  ax = fig.add_subplot(2,2, i+1)
  ax.scatter(derivative, y, color='red')
  ax.plot(x, x_square)
  ax.title.set_text('Learning Rate = '+ str(lr))
```

인공신경망 만들기(ANN)

```python
# 데이터
X = np.array([[0,0], [1,0], [0,1], [1,1]])
y = np.array([[0], [1], [1], [1]])
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.losses import mse
```

```python
# 모델 만들기
# 인공신경망 모형 구성
model = Sequential() # layer를 쌓을 수 있는 판을 sequential이라고 함
model.add(Dense(3, input_shape=(2,), activation= 'linear')) # 은닉층 설정(input_shape으로 입력층 설정함)
model.add(Dense(1, activation= 'linear')) # 출력층 설정 # activation은 활성화 함수를 결정해주는 것

# 인공신경망 모형 컴파일
# optimizer가 경사하강법 종류를 정하는 것(보통 Adam을 씀)
# loss는 손실함수 지정
# metrics는 평가지표 지정
model.compile(optimizer = 'SGD', loss=mse, metrics=['acc'])

# 인공신경망 학습
# 학습시킬 때 learning rate와 epoch를 설정함
model.fit(X, y, epochs=10)
```

iris 데이터 딥러닝 모델 만들기

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
x = iris['data']
```

```python
x.shape
```

```python
y = iris['target']
```

```python
y
```

```python
# iris 데이터 딥러닝 모델 만들기
import tensorflow as tf
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y,
                                                    test_size=0.4,
                                                    random_state=2)
```

```python
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
# onehot 인코딩
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y_train = tf.keras.utils.to_categorical(y_train)
y_test = tf.keras.utils.to_categorical(y_test)
```

```python
y_train.shape
```

```python
# 모델 Sequential 생성
model = tf.keras.Sequential()
# Input Layer 생성
model.add(tf.keras.layers.Flatten(input_shape=(4,)))
# Hidden Layer 생성
model.add(tf.keras.layers.Dense(256, activation='relu'))
model.add(tf.keras.layers.Dense(128, activation='relu'))
model.add(tf.keras.layers.Dense(64, activation='relu'))
model.add(tf.keras.layers.Dense(32, activation='relu'))
model.add(tf.keras.layers.Dense(16, activation='relu'))
model.add(tf.keras.layers.Dense(8, activation='relu'))
# Output Layer 생성
model.add(tf.keras.layers.Dense(3, activation='softmax'))
```

```python
# Model Compile
model.compile(optimizer='nadam',
              loss=tf.keras.losses.CategoricalCrossentropy(),
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
# 모델 학습
history = model.fit(X_train, y_train, epochs=30)
```

```python
# 모델 평가
model.evaluate(X_test, y_test)
```

```python
# 평가된 지표 확인
dict(zip(model.metrics_names, model.evaluate(X_test, y_test)))
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'], label='Loss')
plt.plot(history.history['accuracy'], label='ACC')
plt.xlim(-1, 30)
plt.title('Loss')
plt.legend()
plt.show()
```
