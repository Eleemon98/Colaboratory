```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 데이터 프레임 불러오기
df = pd.read_csv('BIG FIVE 1995-2019.csv')
df.head()
```

```python
df = df.drop('Round', axis=1)
df.head()
```

```python
df = df.drop('Country', axis=1)
df.head()
```

```python
df.columns
```

```python
df['Score'] = df['FT Team 1'] - df['FT Team 2']
df['HScore'] = df['HT Team 1'] - df['HT Team 2']
```

```python
df.head()
```

```python
df = df[['Date', 'Team 1','Team 2', 'Year', 'Score', 'HScore']]
```

```python
df
```

```python
df.info()
```

```python
df.columns
```

```python
df_ob = df[['Date', 'Team 1', 'Team 2']]
df_int = df[['Year', 'Score', 'HScore']]
```

```python
# onehot, incording : 문자로 된 데이터를 숫자와 비슷한 형태로 변경(데이터 전처리)
df_onehot = pd.get_dummies(df_ob)
```

```python
df_total = pd.concat([df_int, df_onehot], axis=1)
```

```python
df_total
```

```python
y = df_total.pop('Score')
```

```python
X = df_total
```

```python
y
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
# 데이터를 나눠주는 모듈 사용
(X_train, X_test, y_train, y_test) = train_test_split(X, y, random_state=1, test_size=0.25)
```

```python
print(X_train.shape, y_train.shape) # X데이터들의 컬럼끼리 맞는지 확인
print(X_test.shape, y_test.shape) # y데이터들은 X 데이터들의 행과 맞는지 확인
```

```python
from sklearn.linear_model import LinearRegression # 모델을 가져온 것
```

```python
lrm = LinearRegression()
```

```python
lrm.fit(X_train, y_train) # 모델에게 데이터 주고 학습시키는 것
```

```python
np.round(lrm.predict(X_test)[6], 0)
```
