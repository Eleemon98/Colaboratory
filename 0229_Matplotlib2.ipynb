막대그래프

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]
values2 = [150, 300, 850]
colors = ['y', 'dodgerblue', 'C2']

plt.bar(x-0.2, values, color = colors, width=0.4) # 막대그래프는 x축과 y축을 같이 입력해줘야함
plt.bar(x+0.2, values2, width=0.4)
plt.xticks(x, years)
plt.show()
# 두 개의 데이터를 같이 비교하고 싶을 때 사용
```

```python
y = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]
values2 = [-150, -300, -850]
colors = ['y', 'dodgerblue', 'C2']

plt.barh(y, values, color=colors, height=0.4)
plt.barh(y, values2, height=0.4)
plt.yticks(y, years)
plt.xticks([-750, -500, -250, 0, 250, 500, 750], [750, 500, 250, 0, 250, 500, 750])
plt.show()
```

> **실습9.
다음 조건에 맞춰 막대그래프를 그리시오.**
>
- 그림과 같은 데이터를 생성하시오.
- 막대의 두께를 0.8로 하시오.
- 막대의 색상은 빨강, 녹색, 파랑, 보라
- before를 수직방향, after를 수평방향으로 생성하여 하나의 그래프에서 2개의 axes가 되도록 출력하시오.
- 각 axes의 title은 ‘Before’와 ‘After’로 설정하고 x 축의 이름을 “ID”, y축의 이름을 “Weight”로 지정하시오.(figuer title은 “Squat”으로 설정)
- befoer와 after를 하나의 그래프에 같이 표현되도록하고 막대 위치는 edge를 설정하고 각각 노란색과 오랜지색으로 표현하시오. (title은 “Squat” , x : “ID”,  y : “Weight”)
- 범례를 왼쪽 위에 생성하시오.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

ids = pd.Series(['m01','m02','m03','m04'])
before = pd.Series([100, 110, 150, 90])
after = pd.Series([160, 130, 170, 120])
df = pd.concat([ids, before, after], axis=1)
df.columns = ['ID', 'Before', 'After']
```

```python
df
```

```python
fig = plt.Figure()

axes1 = plt.subplot(2,1,1)
axes2 = plt.subplot(2,1,2)

plt.suptitle('Squat')

axes1.set_title('Before')
axes2.set_title('After')
axes1.set_xlabel('ID')
axes1.set_ylabel('Weight')

axes1.bar(df['ID'], df['Before'], width=0.8, color=['r', 'g', 'b', 'purple'])
axes2.barh(df['ID'], df['After'], height=0.8, color=['r', 'g', 'b', 'purple'])
# ID 순서를 바꿔야 할 때는 df의 순서를 sort_index(ascending=False)로 넣어주면 됨

fig.tight_layout()
plt.show()
```

```python
plt.bar(df.index, df['Before'], width=0.4, color='yellow', align='edge', label='Before')
plt.bar(df.index+0.4, df['After'], width=0.4, color='orange', align='edge', label='After')

plt.title('Squat')
plt.xlabel('ID')
plt.ylabel('Weight')
plt.xticks(df.index, df.ID)

plt.legend(loc=2)
plt.show()
```

> 실습 10.
다음 데이터를 통해서 산점도 그래프를 출력하시오.
>
- xlabel은 ‘Longitude’, ylabel은 ‘Latitude’, 그래프의 제목은 ‘Dense’
- 마커의 크기는 p_dense의 0.2를 곱한값으로 설정
- 각 마커의 색깔은 0~7까지 순서대로 입력
- cmap은 ‘spring’ 설정
- **plt.text(`x축 위치`, `y축 위치`, `출력할 text`)**
- 각 막커별 city를 입력
- 투명도는 0.5

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dic = {'city': ['Seoul','Incheon','Dejeon','Degu','Ulsan','Busan','Kawngju'],
       'lat':[37.56, 37.45, 36.35, 35.87, 35.53, 35.18, 35.16], # 경도
       'lon':[126.97, 126.70, 127.38, 128.60, 129.31, 129.07, 126.85], # 위도
       'p_dense': [16154, 2751, 2839, 2790, 1099, 4454, 2995]}

df = pd.DataFrame(dic)
df
```

```python
plt.scatter(df['lon'], df['lat'], s=df['p_dense']*0.2, c=df.index, alpha=0.5, cmap='spring')
plt.title('Dense')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# text 입력
for i in df.index:
  plt.text(df['lon'][i], df['lat'][i], df['city'][i])

plt.colorbar()
plt.show()
```
