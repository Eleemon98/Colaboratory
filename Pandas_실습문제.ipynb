- raito(접종률)이 최대값이 100%가 넘는 데이터를 삭제하시오.
- 다음 데이터의 국가별 raito(접종률) 최대값을 구하시오.
- 상위 10개국과 하위 10개국 접종률(ratio)평균을 구하고 차이를 소수점 첫째자리까지 출력

```python
import pandas as pd
import numpy as np

df = pd.read_csv('covid-vaccination-vs-death_ratio.csv', index_col=0)
df.head()
```

```python
df.info()
```

```python
df.drop(df[df['ratio']>100].index, inplace=True)
```

```python
df
```

```python
df.groupby(['country']).max()
```

```python
df_top = df.sort_values('ratio', ascending=False).head(10)
df_bottom = df.sort_values('ratio', ascending=False).tail(10)

df_top_mean = df_top.ratio.mean()
df_bottom_mean = df_bottom.ratio.mean()

result = abs(df_top_mean - df_bottom_mean)

print(round(result, 1))
```
