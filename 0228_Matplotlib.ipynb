```python
import matplotlib.pyplot as plt
dataset_1 = anscombe[anscombe['dataset'] == 'I']
plt.plot(dataset_1['x'], dataset_1['y'])
plt.show()
```

```python
dataset_1 = anscombe[anscombe['dataset'] == 'I']
dataset_2 = anscombe[anscombe['dataset'] == 'II']
dataset_3 = anscombe[anscombe['dataset'] == 'III']
dataset_4 = anscombe[anscombe['dataset'] == 'IV']

fig = plt.figure() # 그래프 기본 틀 생성
axes1 = fig.add_subplot(2, 2, 1) # row, col, index
axes2 = fig.add_subplot(2, 2, 2)
axes3 = fig.add_subplot(2, 2, 3)
axes4 = fig.add_subplot(2, 2, 4)

axes1.plot(dataset_1['x'], dataset_1['y'], 'ro') # red circle
axes2.plot(dataset_2['x'], dataset_2['y'], 'go') # green circle
axes3.plot(dataset_3['x'], dataset_3['y'], 'bo') # blue circle
axes4.plot(dataset_4['x'], dataset_4['y'], 'yo') # yellow circle

axes1.set_title("dataset_1") # 각 데이터 제목 지정
axes2.set_title("dataset_2")
axes3.set_title("dataset_3")
axes4.set_title("dataset_4")
fig.suptitle("Anscombe Data") # 전체 데이터 제목 지정

fig.tight_layout() # 크기에 맞는 데이터 레이아웃 지정
```

Pie Chart(원 그래프)
* 선거 같은 비율을 시각화하기 좋은 그래프
* 5개~6개 정도의 카테고리 안에서 비율이 크게 차이날 때

```python
import matplotlib.pyplot as plt
import pandas as pd

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

df = pd.DataFrame({'labels': labels,
                   'ratio': ratio})
df
```

```python
plt.pie(df.ratio, labels=df.labels, autopct='%.0f',
        colors=['aquamarine', 'chartreuse', 'blueviolet', 'tomato'],
        startangle=270, counterclock=False,
        explode=[0.02, 0.02, 0.1, 0.02],
        wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth':2})
plt.legend(loc=2)
plt.show()
```

- Numpy :  2/7, 오렌지색
- Pandas : 2/7,  노란색
- Python: 3/7, 초록색
- 소수점 둘째자리까지 표현
- 시작 위치는 45도
- 데이터 순서 시계 반대 방향
- 범례표시 : 왼쪽 아래(3)

```python
import matplotlib.pyplot as plt
import pandas as pd

ratio = [2/7, 2/7, 3/7] # 값만 넣어도 알아서 비율 계산함
labels = ['Numpy', 'Pandas', 'Python']

plt.pie(ratio, labels=labels, autopct='%.2f%%',
        colors=['orange', 'yellow', 'green'],
        startangle=45, counterclock=True)
plt.legend(loc=3)
plt.show()
```

```python
import matplotlib.pyplot as plt

weight = [68, 81, 64, 56, 78, 74, 61, 77, 66, 68, 59, 71,
          80, 59, 67, 81, 69, 73, 69, 74, 70, 65]

plt.hist(weight, cumulative=True, label='cumulative=True') # cumulative는 누적값 표현
plt.hist(weight, cumulative=False, label='cumulative=False')
plt.legend(loc='upper left')
plt.show()
```

```python
import matplotlib.pyplot as plt

weight = [68, 81, 64, 56, 78, 74, 61, 77, 66, 68, 59, 71,
        80, 59, 67, 81, 69, 73, 69, 74, 70, 65]
weight2 = [52, 67, 84, 66, 58, 78, 71, 57, 76, 62, 51, 79,
        69, 64, 76, 57, 63, 53, 79, 64, 50, 61]

plt.hist((weight, weight2), histtype='bar')
plt.title('histtype - bar')
plt.figure()

plt.hist((weight, weight2), histtype='barstacked')
plt.title('histtype - barstacked')
plt.figure()

plt.hist((weight, weight2), histtype='step')
plt.title('histtype - step')
plt.figure()

plt.hist((weight, weight2), histtype='stepfilled')
plt.title('histtype - stepfilled')
plt.show()
```
