DataFrame 병합

```python
# merge를 사용하려면 key가 필요함
left = pd.DataFrame(
     {
         "key": ["K0", "K1", "K2", "K3"],
         "A": ["A0", "A1", "A2", "A3"],
         "B": ["B0", "B1", "B2", "B3"],
    }
 )

right = pd.DataFrame(
     {
        "key": ["K0", "K1", "K2", "K3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    }
 )
```

```python
left
```

```python
right
```

```python
pd.merge(left, right) # 공통되는 열의 이름을 on='공통되는 열의 이름' 넣는다
```

```python
left = pd.DataFrame(
     {
        "key1": ["K0", "K0", "K1", "K2"],
        "key2": ["K0", "K1", "K0", "K1"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
    }
 )

right = pd.DataFrame(
     {
         "key1": ["K0", "K1", "K1", "K2"],
         "key2": ["K0", "K0", "K0", "K0"],
         "C": ["C0", "C1", "C2", "C3"],
         "D": ["D0", "D1", "D2", "D3"],
     }
 )
```

```python
left
```

```python
right
```

```python
# inner join은 교집합으로 결합시키는 방식
pd.merge(left, right, on=['key1', 'key2'], how='inner')
```

```python
# outer join은 합집합으로 결합시키는 방식
pd.merge(left, right, on=['key1', 'key2'], how='outer')
```

```python
# left join은 왼쪽 DataFrame에 있는 key만 사용함
pd.merge(left, right, on=['key1', 'key2'], how='left')
```

```python
# right join은 오른쪽 DataFrame에 있는 key만 사용함
pd.merge(left, right, on=['key1', 'key2'], how='right')
```

```python
# concat은 join 없이 합성하는 방식
df1 = pd.DataFrame(
     {
         "A": ["A0", "A1", "A2", "A3"],
         "B": ["B0", "B1", "B2", "B3"],
         "C": ["C0", "C1", "C2", "C3"],
         "D": ["D0", "D1", "D2", "D3"],
     },
     index=[0, 1, 2, 3],
 )

df2 = pd.DataFrame(
     {
         "A": ["A4", "A5", "A6", "A7"],
         "B": ["B4", "B5", "B6", "B7"],
         "C": ["C4", "C5", "C6", "C7"],
         "D": ["D4", "D5", "D6", "D7"],
     },
     index=[4, 5, 6, 7],
 )

df3 = pd.DataFrame(
     {
         "A": ["A8", "A9", "A10", "A11"],
         "B": ["B8", "B9", "B10", "B11"],
         "C": ["C8", "C9", "C10", "C11"],
         "D": ["D8", "D9", "D10", "D11"]     },
     index=[8, 9, 10, 11],
 )
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
pd.concat([df1, df2, df3]) # 기본 행방향으로 합쳐짐
```

```python
pd.concat([df1, df2, df3], axis=1) # concat은 key가 index이기 때문에 컬럼으로 합칠 때는 index를 맞춰야 함
```
