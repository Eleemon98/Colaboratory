```python
import pandas as pd
import numpy as np
import seaborn as sns

df = sns.load_dataset('tips')
```

```python
df.shape
```

```python
df.info()
```

```python
# tip 예측 모델
df['tip']
```

```python
# 독립변수와 종속변수를 나눠야함
# y : tip x : 나머지
y = df.pop('tip')
X = df
```

```python
X
```

```python
# 모델 불러오기
from sklearn.linear_model import LinearRegression
lrm = LinearRegression()
```

```python
# 글자를 숫자로 바꾸기(인코딩하기)
# LabelEncoder는 데이터프레임 자체를 변경시키지 못하고 컬럼 하나씩만 가능함
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

encoder = LabelEncoder()
cols = ['sex', 'smoker', 'day', 'time']
X_cat = pd.DataFrame()
for i in cols:
  X_cat[i] = encoder.fit_transform(X[i])

X_cat
```

```python
# 숫자의 크기 조정(Scaling)
scaler = MinMaxScaler()

X_num = scaler.fit_transform(X[['total_bill', 'size']])
X_num = pd.DataFrame(X_num, columns=['total_bill', 'size'])
X_num
```

```python
# 다시 하나로 합치기
X = pd.concat([X_cat, X_num], axis=1)
```

```python
# 학습 데이터와 평가 데이터 분리
# test_size는 0.25가 기본값
# random_state는 고정된 값에 의해서 지정된 랜덤 값만 뽑힙
from sklearn.model_selection import train_test_split
(X_train, X_test, y_train, y_test) = train_test_split(X, y,
                                                      test_size=0.3,
                                                      random_state = 1)
```

```python
print(X_train.shape, X_test.shape) # 같은 X, y끼리는 열이 맞는지 보고
print(y_train.shape, y_test.shape) # train 끼리, test끼리는 행이 맞는지 봐야함
```

```python
# 모델 학습
lrm.fit(X_train, y_train)
```

```python
pred = lrm.predict(X_test) # 예측값 생성
pred
```

```python
y_test
```

```python
# 평가지표 확인하기
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error, r2_score
from math import sqrt

print('MAE: ', mean_absolute_error(y_test, pred))
print('MSE: ', mean_squared_error(y_test, pred))
print('RMSE: ', sqrt(mean_squared_error(y_test, pred)))
print('MAPE: ', mean_absolute_percentage_error(y_test, pred))
print('r2 Score: ', r2_score(y_test, pred))
```

Input features in order:
1) CRIM: per capita crime rate by town
2) ZN: proportion of residential land zoned for lots over 25,000 sq.ft.
3) INDUS: proportion of non-retail business acres per town
4) CHAS: Charles River dummy variable (1 if tract bounds river; 0 otherwise)
5) NOX: nitric oxides concentration (parts per 10 million) [parts/10M]
6) RM: average number of rooms per dwelling
7) AGE: proportion of owner-occupied units built prior to 1940
8) DIS: weighted distances to five Boston employment centres
9) RAD: index of accessibility to radial highways
10) TAX: full-value property-tax rate per $10,000 [$/10k]
11) PTRATIO: pupil-teacher ratio by town
12) B: The result of the equation B=1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
13) LSTAT: % lower status of the population

Output variable:
1) MEDV: Median value of owner-occupied homes in $1000's [k$]
