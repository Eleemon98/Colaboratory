```python
# 동영상 파일 읽어오기
import cv2

video = 'image/tiger.mp4' # 파일 위치

cap = cv2.VideoCapture(video) # 영상 파일을 읽어서 변수에 할당

if cap.isOpened():
    while True:
        ret, img = cap.read() # ret: 성공여부, img: 이미지 파일
        if ret:
            cv2.imshow(video, img) # 창 이름, 보여주는 이미지
            cv2.waitKey(25) # 이미지 fps 조정 (25ms 지연)
        else:
            break # 영상이 끝났을 때
else:
    print('Can not read Video') # 캡처 객체가 초기화 실패 했을 때

cap.release() # 캡처 close
cv2.destroyAllWindows() 
```

```python
# 웹캠으로부터 영상 읽기
import cv2

cap = cv2.VideoCapture(0) # 보통 0번카메라로 장치되어있을 것임.

if cap.isOpened():
    while True:
        ret, img = cap.read()
        if ret:
            cv2.imshow('camera', img)
            if cv2.waitKey(1) != -1:
                break
        else:
            print('No Frame')
            break
else:
    print('Can not connected Camera')

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

cap = cv2.VideoCapture(0) # 보통 0번카메라로 장치되어있을 것임.

if cap.isOpened():
    while True:
        ret, img = cap.read()
        if ret:
            cv2.imshow('camera', img)
            if cv2.waitKey(1) != -1:
                cv2.imwrite('image/photo.jpg', img) # 웹캠으로 사진찍기
                break
        else:
            print('No Frame')
            break
else:
    print('Can not connected Camera')

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# h: 512, W: 512, C: RGB 전부 검은색의 이미지 생성
img = np.zeros((512,512,3), np.uint8)

# cv2.line(이미지 파일, start위치(x, y), end위치(x, y), color(B, G, R) 0~255, 선의 굵기(ticknees=5))
# lineType : cv2.line_4, cv2.line_8, cv2.line_AA
cv2.line(img, (0, 0), (511, 511), (255, 255, 255), 5)
cv2.line(img, (511, 511), (0, 0), (255, 255, 255), 5)

cv2.imshow("line", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
## 흰색 배경 512 * 512 컬러 이미지에
## 시작점이 100, 100 에서 끝나는 지점이 300, 300일 때
## 빨간색 선(두께는 5)을 그리시오.

import cv2
import numpy as np

img = np.full((512, 512, 3), 255, np.uint8)
# full 대신 img + 255로도 가능함

cv2.line(img, (100, 300), (300, 300), (0, 0, 255), 5)
cv2.line(img, (100, 100), (300, 100), (0, 0, 255), 5)
cv2.line(img, (300, 100), (300, 300), (0, 0, 255), 5)
cv2.line(img, (100, 100), (100, 300), (0, 0, 255), 5)

cv2.imshow("line", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def img_show(img):
    import cv2
    cv2.imshow("image", img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

img = np.full((512, 512, 3), 255, np.uint8)
# full 대신 img + 255로도 가능함

cv2.line(img, (100, 100), (300, 300), (0, 0, 255), 5, cv2.LINE_4)
cv2.line(img, (200, 200), (400, 400), (0, 255, 255), 5, cv2.LINE_8)
cv2.line(img, (100, 0), (100, 100), (255, 0, 255), 5, cv2.LINE_AA)

img_show(img)
```

```python
img = np.full((512, 512, 3), 255, np.uint8)

# 사각형 그리기
cv2.rectangle(img, (50, 50), (150, 150), (0, 125, 125), 5)
cv2.rectangle(img, (50, 50), (150, 150), (0, 125, 125), -1) # 안에 색을 채울 때

img_show(img)
```

```python
img = np.full((512, 512, 3), 255, np.uint8)

# 다각형 그리기(배열에 꼭지점 좌표를 더 넣으면 다각형이 됨)
poly1 = np.array([[[350, 50], [250, 200], [450, 200]]], dtype=np.int32)

# True를 False로 바꾸면 한쪽 면이 열려있는 다각형이 됨
cv2.polylines(img, poly1, True, (0,0,255), 5)

img_show(img)
```

```python
# 원 그리기
img = np.full((512, 512, 3), 255, np.uint8)

cv2.circle(img, (256, 256), 50, (255, 255, 0), -1)
img_show(img)

# 타원 그리기(이미지, 중심, 타원의 각 반지름, 타원의 기울기각도, 시작 각도, 종료 각도, 색상, 두께)
cv2.ellipse(img, (125, 125), (75, 50), 45, 0, 360, (125, 255, 0), -1)
img_show(img)
```

```python
# 이벤트 처리
import cv2

title = "Circle Maker"
img = np.full((512, 512, 3), 255, np.uint8)
cv2.imshow(title, img)

def clickMouse(event, x, y, flags, param):    
    if event == cv2.EVENT_LBUTTONDOWN:
        print(event, x, y)
        cv2.circle(img, (x, y), 30, (0, 0, 0), -1)
        cv2.imshow(title, img)

cv2.setMouseCallback(title, clickMouse)

while True:
    if cv2.waitKey(0) & 0xFF == 27: # 27은 esc 버튼 번호
        break
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import random

title = "Circle Maker"
img = np.full((512, 512, 3), 255, np.uint8)
cv2.imshow(title, img)

colors = {'black': (0, 0, 0),
          'red': (0, 0, 255),
          'blue': (255, 0, 0),
          'green': (0, 255, 0),
          'yellow': (0, 255, 255),
          'cyan': (255, 255, 0),
          'magenta': (255, 0, 255),
          'orange': (0, 165, 255),
          'purple': (128, 0, 128),
          'pink': (147, 20, 255)}

def onMouse(event, x, y, flags, param):
    color = colors['black']
    size = 30
    if event == cv2.EVENT_LBUTTONDOWN:
        print(event, x, y)
        if flags & cv2.EVENT_FLAG_SHIFTKEY and flags & cv2.EVENT_FLAG_CTRLKEY:
            ran_color = random.choice(list(colors.keys()))
            color = colors[ran_color]
            cv2.rectangle(img, (x, y), (x+100, y+100), color, -1)
            cv2.imshow(title, img)
        elif flags & cv2.EVENT_FLAG_CTRLKEY:
            ran_color = random.choice(list(colors.keys()))
            color = colors[ran_color]
            cv2.circle(img, (x, y), size, color, -1)
            cv2.imshow(title, img)
        elif flags & cv2.EVENT_FLAG_SHIFTKEY:
            ran_color = random.choice(list(colors.keys()))
            color = colors[ran_color]
            size = 15
            cv2.circle(img, (x, y), size, color, -1)
            cv2.imshow(title, img)   
        else:
            cv2.circle(img, (x, y), size, color, -1)
            cv2.imshow(title, img)

cv2.setMouseCallback(title, onMouse)

while True:
    if cv2.waitKey(0) & 0xFF == 27:
        break
cv2.destroyAllWindows()
```
