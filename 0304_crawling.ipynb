```python
url = 'https://www.nintendo.co.kr/switch/ranking/ranking_2023_full.html'
res = get_html(url)
```

```python
soup = bs(res, 'html.parser')
soup
```

```python
games = [g.text for g in soup.select('p.tit')]
games
```

```python
date_elem = [d.text for d in soup.select('.releaseInfo')]
date_elem
```

```python
def extract_date(text):
    match = re.search(r'\d{4}\. \d{1,2}\. \d{1,2}', text)
    if match:
        return match.group()
    else:
        return None

# 각 문자열에서 날짜 부분을 추출하여 리스트에 저장
dates = [extract_date(text) for text in date_elem]

# 결과 출력
print(dates)
```

```python
def extract_text(text):
    match = re.search(r'\d{4}\. \d{1,2}\. \d{1,2}', text)
    if match:
        return text[:match.start()] + text[match.end():]
    else:
        return text

# 각 문자열에서 날짜를 제외한 나머지 부분을 추출하여 리스트에 저장
companys = [extract_text(text) for text in date_elem]

print(companys)
```

```python
dic = {'game': games, 'date': dates, 'company': companys}
```

```python
df = pd.DataFrame(dic)
df
```
