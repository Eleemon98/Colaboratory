동영상 파일 불러오기

```python
import cv2

video = '/content/tiger.mp4'

cap = cv2.VideoCapture(video)

res = (360, 240)
fourcc = cv2.VideoWriter_fourcc(*'MP4V')
out = cv2.VideoWriter(video, fourcc, 25, res)

if cap.isOpened():
  while True:
    ret, img = cap.read() # ret: 성공여부, img: 이미지 파일
    if ret:
      cv2_imshow(video, img)
      out.write(img) # 창 이름, 보여주는 이미지
      cv2.waitKey(25) # 이미지 fps 조정 (25ms 지연)
    else:
      break # 영상이 끝났을 때
else:
  print('Can not read Video') # 캡처 객체가 초기화 실패 했을 때

cap.release() # 캡처 close
cv2.destroyAllWindows()
```

```python
import cv2

video_path = '/content/tiger.mp4'

# 비디오 파일 열기
cap = cv2.VideoCapture(video_path)

# 비디오 파일을 열 수 있는지 확인
if not cap.isOpened():
    print("비디오를 열 수 없습니다.")
else:
    # 비디오의 속성 가져오기
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))

    # 비디오 저장을 위한 VideoWriter 객체 초기화
    fourcc = cv2.VideoWriter_fourcc(*'MP4V')
    out = cv2.VideoWriter('/content/tiger_output.mp4', fourcc, fps, (frame_width, frame_height))

    while cap.isOpened():
        ret, frame = cap.read()  # 프레임 읽기
        if ret:
            # 프레임 출력
            cv2.imshow('Video', frame)
            out.write(frame)  # 비디오에 프레임 쓰기
            if cv2.waitKey(25) & 0xFF == ord('q'):
                break  # 'q' 키를 누르면 종료
        else:
            break  # 비디오가 끝나면 종료

    # 비디오 파일 닫기
    cap.release()
    out.release()
    cv2.destroyAllWindows()
```
